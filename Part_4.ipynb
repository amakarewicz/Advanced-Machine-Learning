```python
# imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis as QDA
from sklearn.neighbors import KNeighborsClassifier

from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

```python
# reading preprocessed datasets
data_adult_train = pd.read_csv('datasets/project1/preprocessed/adult_train_x.csv')
data_adult_test = pd.read_csv('datasets/project1/preprocessed/adult_test_x.csv')

data_credit_train = pd.read_csv('datasets/project1/preprocessed/credit_train_x.csv')
data_credit_test = pd.read_csv('datasets/project1/preprocessed/credit_test_x.csv')

data_sick_train = pd.read_csv('datasets/project1/preprocessed/sick_train_x.csv')
data_sick_test = pd.read_csv('datasets/project1/preprocessed/sick_test_x.csv')

data_titanic_train = pd.read_csv('datasets/project1/preprocessed/titanic_train_x.csv')
data_titanic_test = pd.read_csv('datasets/project1/preprocessed/titanic_test_x.csv')
```

```python
# reading preprocessed datasets
target_adult_train = pd.read_csv('datasets/project1/preprocessed/adult_train_y.csv').target.to_numpy()
target_adult_test = pd.read_csv('datasets/project1/preprocessed/adult_test_y.csv').target.to_numpy()

target_credit_train = pd.read_csv('datasets/project1/preprocessed/credit_train_y.csv').target.to_numpy()
target_credit_test = pd.read_csv('datasets/project1/preprocessed/credit_test_y.csv').target.to_numpy()

target_sick_train = pd.read_csv('datasets/project1/preprocessed/sick_train_y.csv').target.to_numpy()
target_sick_test = pd.read_csv('datasets/project1/preprocessed/sick_test_y.csv').target.to_numpy()

target_titanic_train = pd.read_csv('datasets/project1/preprocessed/titanic_train_y.csv').target.to_numpy()
target_titanic_test = pd.read_csv('datasets/project1/preprocessed/titanic_test_y.csv').target.to_numpy()
```

```python
lda = LDA()
lda.fit(data_adult_train, target_adult_train)
```

```python
accuracy_score(target_adult_test, lda.predict(data_adult_test))
```

```python
qda = QDA()
qda.fit(data_adult_train, target_adult_train)
```

```python
accuracy_score(target_adult_test, qda.predict(data_adult_test))
```

```python
corr = data_adult_train.corr()
plt.figure(figsize=(40,30))
sns.heatmap(corr, annot=True) # collinear features after one-hot-encoding
```

```python
knn = KNeighborsClassifier(n_neighbors=20)
knn.fit(data_adult_train, target_adult_train)
```

```python
accuracy_score(target_adult_test, knn.predict(data_adult_test))
```
